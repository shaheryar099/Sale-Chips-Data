## Performing EDA

```python
df1['PROD_NAME'].value_counts().nunique() #There are 101 total product. 
```

```python
df1['BRAND'].value_counts().sort_values().plot(kind = 'barh', label = 'Count')#Most opted brand is KETTLE and then Woolworths
plt.title('Value count of Brand')
plt.legend()
plt.show()
```

```python
plt.pie(x = df1['LIFESTAGE'].value_counts().values, 
        labels=df1['LIFESTAGE'].value_counts().index,
       autopct='%.2f%%')
plt.show()
```

### After Kettle most opted brand is smiths.

```python
plt.title('Count of customer')
ax = sns.countplot(x = df1['PREMIUM_CUSTOMER'])
ax.bar_label(ax.containers[0])
```

### Most business given by mainstream and budget customers then premium customer giving business

```python
r1 = df1['PROD_NAME'].value_counts().head(15).sort_values()
r1.plot(kind = 'barh')
```

### "Kettle Sweet Chilli And Sour Cream 175g" is most opted project by customer    

```python
df1.columns
```

```python
r2 = df1.groupby('PROD_NAME').agg(PROD_QTY_SUM=('PROD_QTY', 'sum'),sales_sum = ('TOT_SALES','sum')).reset_index().sort_values(by = 'sales_sum', ascending=False)
r2.head(15)
```

### "Dorito Corn Chp Supreme 380g" gives most sales

```python
r2 = df1.groupby('PROD_NAME').agg(PROD_QTY_SUM=('PROD_QTY', 'sum'),
                                  sales_sum = ('TOT_SALES','sum')).reset_index().sort_values(by = 'PROD_QTY_SUM', ascending=False)
r2.head(15)
```

### "Kettle Mozzarella Basil & Pesto 175g" is most qty ordered product.

```python
r3 = df1.groupby('BRAND')[['TOT_SALES']].sum().reset_index().sort_values(by = 'TOT_SALES', ascending = False)
r3
```

```python
plt.title('Total sales Vs Brand')
sns.barplot(data = r3, x = 'BRAND', y = 'TOT_SALES')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.title('Total Sales Vs lifestage & customer')
sns.lineplot(data = df1, y = 'TOT_SALES', x = 'LIFESTAGE', hue = 'PREMIUM_CUSTOMER', ci = None)
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```
